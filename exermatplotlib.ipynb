```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Importe matplotlib.pyplot como plt e defina% matplotlib inline se você estiver usando o notebook jupyter. Qual comando você usa se você não estiver usando o notebook jupyter?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show()
```

# Exercício 1
**Acompanhe estes passos:**

**Crie um objeto de figura chamado fig usando plt.figure ()**

**Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax".**

**Plote (x, y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo:**

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('Eixo X')
axes.set_ylabel('Eixo Y')
axes.set_title('Título')
```

# Exercício 2
**Crie um objeto de figura e coloque dois eixos sobre ele, ax1 e ax2. Localizado em [0,0,1,1] e [0.2, 0.5, 0.2, 0.2], respectivamente.**

# Exercício 3
**Crie o gráfico abaixo, adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0.2, 0.5, 0.4, 0.4]**

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

**Agora use x, y e z arrays para criar um gráfico para x^2, onde no gráfico interno queremos as informações do eixo x entre os valores de 20 e 22, no eixo y queremos entre 30 e 50 e tenha título de zoom .**

```python
axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom')
axes2.set_xlim(20,22)
axes2.set_xlim(30,50)

fig
```

# Exercício 4
**Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo.**

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
```

**Agora trace (x, y) e (x, z) nos eixos. Use diferentes estilos de linhas e aumente a espessura da linha.**

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(x,y,color='blue', ls='--', lw=3)
ax[1].plot(x,z,color='red', ls='-', lw=3)
```

**Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () apenas copiando e colando seu código anterior.**

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(20,3))
ax[0].plot(x,y,color='blue', ls='-', lw=5)
ax[1].plot(x,z,color='red', ls='--', lw=2)
```
